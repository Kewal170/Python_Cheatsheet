**-------------------- loc & iloc --------------------**

```python
coffee.head()
```

```python
coffee.loc[0] # coffee.loc[rows, column]
```

```python
coffee.loc[[0,1,2,5]]
```

```python
coffee.loc[0:5]
```

```python
coffee.loc[:10]
```

```python
coffee.loc[5:14]
```

```python
coffee.loc[1:5,["Day","Units Sold"]]
```

```python
coffee.iloc[1:6,[0,2]] # only diff btwn loc & iloc => loc uses Column name & iloc uses Column index
```

```python
coffee.loc[1,"Units Sold"] = 10
```

```python
coffee.head()
```

```python
coffee.at[2,"Units Sold"]
```

```python
coffee.Day # coffee["Day"]
```

```python
coffee.sort_values("Units Sold", ascending =False) # descending
```
